## Perfoming encoding using loop on condition

```python
from pandas.core.dtypes.common import is_numeric_dtype, is_categorical_dtype
```

```python
insurance_labelencoded_loop.head()
```

```python
for column in insurance_labelencoded_loop.columns:
    if is_numeric_dtype(insurance_labelencoded_loop[column]):
        continue
    else:
        insurance_labelencoded_loop[column] = led.fit_transform(insurance_labelencoded_loop[column])

insurance_labelencoded_loop.head()
```

## Onehot encoding `sklearn`

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohed = OneHotEncoder(handle_unknown="ignore", drop="first")
```

```python
insurance_onehotencoded["sex"].shape
```

```python
insurance_onehotencoded["sex"].values.reshape(-1,1).shape
```

```python
ohed.fit_transform(insurance_onehotencoded["sex"].values.reshape(-1,1)).toarray()
```

```python
ohed.categories_
```

```python
ohed.fit_transform(insurance_onehotencoded["smoker"].values.reshape(-1,1)).toarray()
```

```python
ohed.categories_
```

```python
ohed.fit_transform(insurance_onehotencoded[["sex", "smoker", "region"]]).toarray()
```

```python
ohed_array = ohed.fit_transform(insurance_onehotencoded[["sex", "smoker", "region"]]).toarray()
ohed_array
```

```python
ohed.categories_
```

```python
import numpy as np
```

```python
features = ohed.categories_
features
```

```python
features_lst = []
for i in features:
    for j in i[1:]:
        features_lst.append(j)
features_lst
```

```python
oh_encoded_features = pd.DataFrame(ohed_array, columns=features_lst)
oh_encoded_features.head()
```

```python
pd.concat([insurance_onehotencoded, oh_encoded_features], axis=1)
```

```python
dropped_insurace_ohed = insurance_onehotencoded.drop(['sex', 'smoker', 'region'], axis=1)
dropped_insurace_ohed.head()
```

```python
pd.concat([dropped_insurace_ohed, oh_encoded_features], axis=1)
```

## Ordinal encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
insurance_ordinal.head()
```

```python
city = insurance_ordinal.region.unique()
city
```

```python
oed = OrdinalEncoder(categories=[city])
```

```python
oed_area = oed.fit_transform(insurance_ordinal[["region"]])
oed_area
```

```python
oed.categories_
```

```python
insurance_ordinal.head(10)
```

```python
insurance_ordinal["region"] = oed_area
insurance_ordinal.head(10)
```
